```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import time


#預設人數 亞洲土豪 暫定人數 
count = "405,389"

#使用chrome的webdriver
browser = webdriver.Chrome()
#開啟網頁
browser.get('http://35.201.209.150:5000/')
#視窗最大化
browser.maximize_window()
#輸入帳號密碼進行登入
element = browser.find_element_by_name ("username").send_keys("tami")
element = browser.find_element_by_name ("pwd").send_keys("tami")
element = browser.find_element_by_css_selector('.btn.btn-primary.hidden-xs.btn-login').click()

#開啟左側特徵管理
time.sleep(3)
element = browser.find_element_by_xpath('//*[@id="menu"]/ul/li[4]/a').click()
#切換到特徵列表
time.sleep(3)
element = browser.find_element_by_xpath('//*[@id="menu"]/ul/li[4]/ul/li[1]/a').click()
#按下新增特徵+
time.sleep(3)
element = browser.find_element_by_xpath('/html/body/section/div/section/div/div[1]/div/div[1]/button').click()
#選擇特徵歸群 1基本輪廓 2關注事項 
time.sleep(5)
element = browser.find_element_by_xpath('//*[@id="cate1"]/option[4]').click()
#按下新增條件
time.sleep(3)
element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/section[1]/header/h2/button').click()
time.sleep(3)
element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/section[1]/div/div/div/section/div/div[1]/div/div/select/option[5]').click()
time.sleep(3)
#資料來源 TDR Tree
#對應下層 tdr_label1
element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/section[1]/div/div/div/section/div/div[2]/div[1]/a').click()

#第一層 亞洲
s1 = Select(browser.find_element_by_id('tdr_label1'))
s1.select_by_visible_text("亞洲")
time.sleep(1)
#第二層 全選
element = browser.find_element_by_xpath('//*[@id="modalTDRTag"]/section/div/div/div/div[1]/div[3]/div[2]/div/div[1]/span/span/span/div/button').click()
element = browser.find_element_by_xpath('//input[@value="multiselect-all"]').click()
time.sleep(3)
element = browser.find_element_by_xpath('//*[@id="modalTDRTag"]/section/div/div/div/div[1]/div[3]/div[2]/div/div[1]/span/span/span/div/button').click()
#新增
element = browser.find_element_by_xpath('//*[@id="modalTDRTag"]/section/footer/div/div/button[1]').click()

#指定時間區間-2019/08/01~2019/11/30
element = browser.find_element_by_name ("start").send_keys("2019/08/01")
element = browser.find_element_by_name ("end").send_keys("2019/11/30")

#輸入特徵命名:亞洲土豪
element = browser.find_element_by_name ("feature_name").send_keys("亞洲土豪")
time.sleep(3)
#輸入特徵描述:亞洲人都是暴發戶
element = browser.find_element_by_name ("feature_description").send_keys("亞洲人都是暴發戶")
#權限設定:開放編輯
element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/div[3]/div[2]/label[2]/input').send_keys(Keys.SPACE)
time.sleep(10)


#取的數字並印出
element1 = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/header/div/div[2]/span')
print ("取得的數字：" , element1.text)
print ("預期的數字：" , count)


#判斷是否數字正確
if(element1.text == count):
    #創建此特徵
    element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/div[5]/div/button[2]').click()
    print("測試案例成功")
else:
    element = browser.find_element_by_xpath('/html/body/section/div/section/section/div/div/div[1]/div[5]/div/button[1]').click()
    print("測試案例失敗")

#如果需要執行完自動關閉，就要加上下面這一行
#browser.close()
```
